```python
import os
import random
import re
import csv
from PIL import Image
import matplotlib.pyplot as plt

diretorio_raiz = os.getcwd()

def extrair_variavel_ground_truth(nome_arquivo):
    nome_arquivo = os.path.basename(nome_arquivo)
    nome_arquivo = os.path.splitext(nome_arquivo)[0]
    nome_base = nome_arquivo.split('_valor_')[0]
    return nome_base.strip()

ground_truth_path = os.path.join(diretorio_raiz, 'ground_truth.csv')
ground_truth = {}

with open(ground_truth_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    next(reader)
    for row in reader:
        if len(row) >= 2:
            chave = row[0].strip()
            valor = row[1].strip()
            ground_truth[chave] = valor

recortes_dir = os.path.join(diretorio_raiz, 'recortes')
imagens = [f for f in os.listdir(recortes_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

total_arquivos = 0
acertos_totais = 0

for img_nome in imagens:
    chave = extrair_variavel_ground_truth(img_nome)
    valor_esperado = ground_truth.get(chave, '')

    match = re.search(r'_valor_(.+)$', img_nome)
    valor_extraido = match.group(1).replace('.jpg', '').replace('.png', '') if match else ''

    if valor_esperado != '':
        total_arquivos += 1
        if valor_extraido.strip() == valor_esperado.strip():
            acertos_totais += 1

if total_arquivos > 0:
    acuracia_total = (acertos_totais / total_arquivos) * 100
    print(f"Acurácia total em {total_arquivos} arquivos: {acuracia_total:.2f}% ({acertos_totais}/{total_arquivos})")
else:
    print("Nenhum arquivo válido para cálculo de acurácia total.")

amostras = random.sample(imagens, min(10, len(imagens)))

acertos = 0
total = 0

for img_nome in amostras:
    img_path = os.path.join(recortes_dir, img_nome)
    chave = extrair_variavel_ground_truth(img_nome)

    if chave in ground_truth:
        valor_esperado = ground_truth[chave].replace("/", '')
        
        print(f"\nArquivo: {img_nome}\nChave: {chave}")
        match = re.search(r'_valor_(.+)$', img_nome)
        valor_extraido = match.group(1).replace('.jpg', '').replace('.png', '') if match else '-'

        correto = valor_extraido.strip() == valor_esperado.strip()
        total += 1
        if correto:
            acertos += 1

        img = Image.open(img_path)
        plt.figure(figsize=(4, 4))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Arquivo: {img_nome}\nExtraído: {valor_extraido}\nEsperado: {valor_esperado}\n{'Correto' if correto else 'Errado'}")
        plt.show()

if total > 0:
    acuracia = (acertos / total) * 100
    print(f"\nAcurácia nas {total} amostras: {acuracia:.2f}% ({acertos}/{total})")
else:
    print("Nenhuma comparação válida foi possível (valores esperados ausentes no ground_truth).")
```
